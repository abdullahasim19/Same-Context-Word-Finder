## Importing the Libraries

```python
import gzip
import gensim 
import logging
import pickle
import re
from gensim.utils import tokenize
from gensim.models import Word2Vec
from collections import Counter, defaultdict


logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)

documents=[]
```

## Loading the scraped data of Hamari Web

```python
#processing the Hamari web data
f = open('hamari_web_afsanay.pkl', 'rb') 
test_dict = pickle.load(f)
f.close()

i=0;
for key, value in test_dict.items():
    value=re.findall(r'[\u0600-\u06ff\u0750-\u077f\ufb50-\ufdff\ufe70-\ufeff]+ ?\n?\t?',value)
    value="".join(value)
    lst=list(tokenize(value))
    documents.append(lst)
    
```

## Loading the scraped data of Urdu Wiki

```python
#processing the urdu wiki data
f = open('urdu_wiki.pkl', 'rb') 
test_dict = pickle.load(f)
f.close()

for key, value in test_dict.items():
    value=re.findall(r'[\u0600-\u06ff\u0750-\u077f\ufb50-\ufdff\ufe70-\ufeff]+ ?\n?\t?',value)
    value="".join(value)
    lst=list(tokenize(value))
    documents.append(lst)
```

## Loading the scraped data of BBC Urdu

```python
#processing the bbc urdu data   
f = open('bbc_urdu.pkl', 'rb') 
test_dict = pickle.load(f)
f.close()

for key, value in test_dict.items():
    value=re.findall(r'[\u0600-\u06ff\u0750-\u077f\ufb50-\ufdff\ufe70-\ufeff]+ ?\n?\t?',value)
    value="".join(value)
    lst=list(tokenize(value))
    documents.append(lst)
```

## Creating Our Word2Vec Model

```python
model = Word2Vec (documents, vector_size=300, window=10, min_count=19, workers=10)
```

## Training Our Word2Vec Model

```python
model.train(documents,total_examples=len(documents),epochs=10)
```

## Saving Our Word2Vec Model

```python
model.save("word2vecFinal1_300d_allData.model")
```
