# Scraping from Urdu Wiki

## Importing the libraries

```python
from bs4 import BeautifulSoup
import lxml
import requests
import pandas as pd
import numpy as np
import pickle
```

```python
urduwiki={}
list_of_headings=[]
```

## Scraping the data

```python
def scrap_all_data(link):
    
    list_of_headings.append(link)
    
    while len(list_of_headings)>0:
        try:
            temp_source=requests.get(list_of_headings.pop(0))
            temp_r=temp_source.content
            temp_soup=BeautifulSoup(temp_r,'html.parser')
            temp_heading=temp_soup.find('h1').get_text()
            
            if urduwiki.get(temp_heading) == None:
                print(temp_heading,end='\t')
                paragraphs=temp_soup.find('div',id='mw-content-text').find_all('p')
                para_text=''
                for temp in paragraphs:
                    para_text+=temp.get_text()

                #populate dictionary
                urduwiki.update({temp_heading:para_text})
                
                #maintaining .pkl file 
                f = open('urdu_wiki.pkl', 'wb')   # 'wb' instead 'w' for binary file
                pickle.dump(urduwiki, f, -1)       # -1 specifies highest binary protocol
                f.close()    
                
                sublink='https://ur.wikipedia.org'

                for i in paragraphs:
                    temp_anchors=i.find_all('a')
                    for j in temp_anchors:
                        try:
                            list_of_headings.append(sublink+j.get('href'))
                        except:
                            v=0 #does nothing
                try:
                    table=temp_soup.find_all('table', class_='sortable wikitable jquery-tablesorter')
                    table_anchors=table.find_all('a')

                    for i in table_anchors:
                        try:
                            list_of_headings.append(sublink+i.get('href'))
                        except:
                            vxx=0 #does nothing

                except:
                    va=0 #does nothing
                    
        
        except:
            vaa=1 #does nothing
```

## Scraping Each Link

```python
url='https://ur.wikipedia.org/w/index.php?title=%D8%B5%D9%81%D8%AD%DB%82_%D8%A7%D9%88%D9%84&action=edit'
source=requests.get(url).text
soup=BeautifulSoup(source, 'lxml')
index = 0
div = soup.find('div', class_='templatesUsed')
link = div.find_all('li')
for li in link:
    print(index)
    print(li.find('a').text)
    link_='https://ur.wikipedia.org' + li.find('a')['href']
    print('\n')
    scrap_all_data(link_)
    index+=1
print('load successfully')
```
